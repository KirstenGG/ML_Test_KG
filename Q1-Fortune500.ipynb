```python
df_2023 = df[df['year'] == 2023]
```

```python
# market value by Industry/Sector

total_market_value = df_2023['market_value_mil'].sum()
print(total_market_value)

market_value_by_industry = df_2023.groupby('industry')['market_value_mil'].sum()
print(market_value_by_industry)

market_value_by_sector = df_2023.groupby('sector')['market_value_mil'].sum()
print(market_value_by_sector)

market_value_percentage_by_sector = (market_value_by_sector / total_market_value) * 100
print(market_value_percentage_by_sector)
```

```python
market_value_by_industry.sort_values(inplace=True)  # Sort values for better visualization
plt.figure(figsize=(16, 14))
market_value_by_industry.plot(kind='barh')  # Horizontal bar plot
plt.title('Total Market Value by Industry for 2023')
plt.xlabel('Total Market Value (in Million)')
plt.ylabel('Industry')
plt.show()
```

```python
market_value_by_sector.sort_values(inplace=True)  # Sort values for better visualization
plt.figure(figsize=(10, 8))
market_value_by_sector.plot(kind='barh')  # Horizontal bar plot
plt.title('Total Market Value by Sector for 2023')
plt.xlabel('Total Market Value (in Million)')
plt.ylabel('Industry')
plt.show()
```

```python
market_value_by_sector.sort_values(inplace=True)  # Sort values for better visualization
total_market_value = market_value_by_sector.sum()  # Calculate total market value

plt.figure(figsize=(10, 8))
bars = plt.barh(market_value_by_sector.index, market_value_by_sector.values)  # Horizontal bar plot

# Add percentage labels to each bar
for bar in bars:
    width = bar.get_width()  # Get the width of each bar
    percentage = (width / total_market_value) * 100  # Calculate percentage
    label_x_pos = width + total_market_value * 0.01  # Position the label slightly right of the bar end
    plt.text(label_x_pos, bar.get_y() + bar.get_height() / 2.0, f'{percentage:.2f}%', va='center')

plt.title('Total Market Value by Sector for 2023')
plt.xlabel('Total Market Value %')
plt.ylabel('Sector')
plt.tight_layout()  # Adjust layout to make room for the percentage labels if necessary
plt.show()
```

```python
# total revenue by Industry/Sector

total_revenue = df_2023['revenue_mil'].sum()
print(total_revenue)

revenue_by_industry = df_2023.groupby('industry')['revenue_mil'].sum()
print(revenue_by_industry)

revenue_by_sector = df_2023.groupby('sector')['revenue_mil'].sum()
print(revenue_by_sector)

revenue_percentage_by_sector = (revenue_by_sector / total_revenue) * 100
print(revenue_percentage_by_sector)
```

```python
revenue_by_sector.sort_values(inplace=True)  # Sort values for better visualization
total_revenue = revenue_by_sector.sum()  # Calculate total market value

plt.figure(figsize=(10, 8))
bars = plt.barh(revenue_by_sector.index, revenue_by_sector.values)  # Horizontal bar plot

# Add percentage labels to each bar
for bar in bars:
    width = bar.get_width()  # Get the width of each bar
    percentage = (width / total_revenue) * 100  # Calculate percentage
    label_x_pos = width + total_revenue * 0.01  # Position the label slightly right of the bar end
    plt.text(label_x_pos, bar.get_y() + bar.get_height() / 2.0, f'{percentage:.2f}%', va='center')

plt.title('Total Revenue by Sector for 2023')
plt.xlabel('Total Revenue (in Million)')
plt.ylabel('Sector')
plt.tight_layout()  # Adjust layout to make room for the percentage labels if necessary
plt.show()
```

```python
# total profit by Industry/Sector

total_profit = df_2023['profit_mil'].sum()
print(total_profit)

profit_by_industry = df_2023.groupby('industry')['profit_mil'].sum()
print(profit_by_industry)

profit_by_sector = df_2023.groupby('sector')['profit_mil'].sum()
print(profit_by_sector)

profit_percentage_by_sector = (profit_by_sector / total_profit) * 100
print(profit_percentage_by_sector)
```

```python
profit_by_sector.sort_values(inplace=True)  # Sort values for better visualization
total_profit = profit_by_sector.sum()  # Calculate total profit

plt.figure(figsize=(10, 8))
bars = plt.barh(profit_by_sector.index, profit_by_sector.values)  # Horizontal bar plot

# Add percentage labels to each bar
for bar in bars:
    width = bar.get_width()  # Get the width of each bar
    percentage = (width / total_profit) * 100  # Calculate percentage
    label_x_pos = width + total_profit * 0.01  # Position the label slightly right of the bar end
    plt.text(label_x_pos, bar.get_y() + bar.get_height() / 2.0, f'{percentage:.2f}%', va='center')

plt.title('Total Profit by Sector for 2023')
plt.xlabel('Total Profit (in Million)')
plt.ylabel('Sector')
plt.tight_layout()  # Adjust layout to make room for the percentage labels if necessary
plt.show()
```

```python
# profit margin by sector

total_profit_margin = total_profit/total_revenue*100
print(total_profit_margin)

profit_margin_by_industry = profit_by_industry/revenue_by_industry*100
print(profit_margin_by_industry)

profit_margin_by_sector = profit_by_sector/revenue_by_sector*100
print(profit_margin_by_sector)
```

```python
profit_margin_by_sector.sort_values(inplace=True)  # Sort values for better visualization
total_profit = profit_by_sector.sum()  # Calculate total market value

plt.figure(figsize=(10, 8))
bars = plt.barh(profit_margin_by_sector.index, profit_margin_by_sector.values)  # Horizontal bar plot

# Add percentage labels to each bar
for bar in bars:
    width = bar.get_width()  # Get the width of each bar
    percentage = width  # Calculate percentage
    plt.text(label_x_pos, bar.get_y() + bar.get_height() / 2.0, f'{percentage:.2f}%', va='center')

plt.title('Profit Margin by Sector for 2023')
plt.xlabel('Profit Margin %')
plt.ylabel('Sector')
plt.tight_layout()  # Adjust layout to make room for the percentage labels if necessary
plt.show()
```

```python
profit_margin_by_industry.sort_values(inplace=True)  # Sort values for better visualization

plt.figure(figsize=(16, 14))
bars = plt.barh(profit_margin_by_industry.index, profit_margin_by_industry.values)  # Horizontal bar plot

# Add labels to each bar
for bar in bars:
    width = bar.get_width()  # Get the width of each bar
    # Assuming width itself is the profit margin percentage
    label_x_pos = width + max(profit_margin_by_industry.values) * 0.01  # Position the label slightly right of the bar end, using 1% of the max value for offset
    plt.text(label_x_pos, bar.get_y() + bar.get_height() / 2.0, f'{width:.2f}%', va='center')

plt.title('Profit Margin by Industry for 2023')
plt.xlabel('Profit Margin %')
plt.ylabel('Industry')
plt.tight_layout()  # Adjust layout to make room for the labels if necessary
plt.show()
```
